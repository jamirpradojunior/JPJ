```python
print(manchete)
```

Exemplo prático para entender a indexação do Python e os "slicings" (fatiamentos)

```python
# Acessar o sexto caractere. Não confundir o número do índice com o algarismo ordinal (ex., nesse caso o caracter '5' representa o sexto caractere)

manchete[5]
```

```python
# Acessar um intervalo de caracteres ":"

# Lembrar que o primeiro é inclusivo e o segundo é exclusivo. Neste exemplo, do caracter 0 até o 8 (não incluindo o 8)

manchete[0:8]
```

```python
# Criar uma variável com os caracteres de 0 até 5 (não incluindo o 5)

acao_do_momento = manchete[0:5]
```

```python
print(acao_do_momento)
```

```python
manchete
```

```python
# Do ínicio até o fim

manchete[:]
```

```python
# Do ínicio até o 10 caractere de trás para frente (sem incluir o -10). Até fixar vale a pena voltar no exemplo prático da documentação Python acima

# Se contar de trás para frente, o -10 é a letra 'd'

manchete[:-10]
```

```python
# Do índice 5 até o -10 (décimo de trás para frente, sem incluir o -10)

manchete[5:-10]
```

```python
# Agora temos um segundo ':'

# Indica de quantos em quantos caracteres devo percorrer o intervalo

# Nesse caso, percorrer todo o intervalo mas indo de -1 em -1 (ou seja, de trás para frente)

manchete[::-1]
```

```python
acao = manchete[0:5]
```

```python
# Todo o intervalo da variável 'acao', porém de trás para frente

acao[::-1]
```

**Resumindo o slicing:**

Uma forma de definirmos as notações string[a:b:c] seria algo como: Execute um comando de fatiamento da string em intervalos de ‘c’, a partir de ‘a’ (incluso) até ‘b’ (não incluso). Se ‘c’ for negativo, contamos de trás para frente. Quando ‘c’ não está presente no código, o padrão é 1 (i.e., fatiar de 1 em 1 caractere). Se ‘a’ não está presente, então você começa o mais distante possível na direção que você está contando (i.e., se ‘c’ é positivo, começa no primeiro caractere. Se ‘c’ é negativo, começa no último caractere). Se ‘b’ não está presente, então você termina o mais distante possível na direção que está contando (i.e., se ‘c’ é positivo, termina no último caractere. Se ‘c’ é negativo, termina no primeiro caractere).

## 17. Kwargs

```python
def compra_de_acoes(**kwargs):
  if 'acao' in kwargs:
    print('A ação comprada foi ', kwargs['acao'])
  else:
    print('O cliente não investe em ações')
```

```python
compra_de_acoes(acao = 'PETR4')
```

```python
compra_de_acoes()
```

```python
def montagem_posicao(*args,**kwargs):
  if 'acao' and 'lote' in kwargs:
    print('O cliente montou uma posição de ', ' e '.join(kwargs['acao']))
    print('Ele comprou ', ' e '.join(args), 'papéis, respectivamente')
    print('Todas elas foram compradas no lote', kwargs['lote'])
```

```python
montagem_posicao('100', '200', acao = ('PETR4', 'VALE3'), lote = 'inteiro')
```
